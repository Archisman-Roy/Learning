Importing the necessary modules

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Required functions

def make_arrays_1D(n, num_arrays):
    '''
    Generate num_arrays number of random walks each of length n (1D)
    '''
    y_arrays = []

    for i in range(num_arrays):
        y = np.zeros(n)

        for i in range(1, n):
            num = np.random.randint(1, 3)
            if num == 1:
                y[i] = y[i-1] + 1
            elif num == 2:
                y[i] = y[i-1] - 1


        y_arrays.append(y)

    return y_arrays


def get_color_array(n):
    '''
    Returns array of colors in random order
    '''
    color_array = ['aquamarine', 'black', 'blue', 'blueviolet', 'brown',
            'chocolate', 'coral', 'crimson', 'cyan',
            'deeppink', 'deepskyblue', 'dodgerblue', 'firebrick',
            'forestgreen', 'fuchsia', 'gainsboro', 'goldenrod',
            'green', 'greenyellow', 'indianred',
            'indigo', 'lawngreen', 'lime',
            'magenta', 'maroon', 'midnightblue', 'mistyrose',
            'olive', 'peru', 'pink', 'plum', 'purple',
            'red', 'rosybrown', 'royalblue', 'saddlebrown',
            'seagreen', 'sienna', 'slateblue',
            'springgreen', 'steelblue', 'tan', 'teal',
            'tomato', 'turquoise', 'violet', 'yellow']

    color_array = np.random.choice(color_array, size = n)
    return color_array



def plot_1D(y_arrays, num_walkers):
    '''
    Plot each array in y_arrays on same plot (1D)
    '''
    colors = get_color_array(num_walkers)
    x = range(len(y_arrays[0]))
    for i in range(num_walkers):
        plt.plot(x, y_arrays[i], alpha=0.7, color=colors[i])
    plt.axis('off')
    plt.show()
    
def Average(lst): 
    return sum(lst) / len(lst) 
```

```python
#Actual Plots
walkers = 10000
timesteps = 500

y_arrays = make_arrays_1D(timesteps, walkers)
plot_1D(y_arrays, walkers)
#Compute average distance 
last_element = []
for i in range(walkers):
    last_element.append(next(reversed(y_arrays[i])))

#The average of the distance from the starting point will be close to zero
print("Average distance from starting point:", Average(last_element))

#However of average of abs(distance) from the starting point will be sq.rt(#timesteps)
abs_last_element =  [abs(ele) for ele in last_element] 
print("Average abs distance from starting point:", Average(abs_last_element))
print("Square root of number of time steps: ", np.sqrt(timesteps))
```
