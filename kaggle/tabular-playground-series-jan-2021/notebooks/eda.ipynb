```python
# imports
import pandas as pd
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# apply settings
sns.set_theme()
%matplotlib inline
```

```python
# read data
train = pd.read_csv("../inputs/train.csv")
train = train.sample(frac=0.01).reset_index(drop=True)
```

```python
# basics
train.shape
```

```python
# keep useful columns
train.drop(["id"], axis=1,inplace = True)
```

```python
# distribution of feature columns
plt.figure(figsize=(24, 12))
sns.boxplot(x="variable", y="value", data=pd.melt(train.drop(['target'], axis=1)))
sns.despine(offset=10, trim=True)
```

```python
# corr of feature fields
sns.pairplot(train[['cont1','cont2','cont3']], corner=True, diag_kind = 'kde', height = 4)
```

```python
# distribution of target variable
sns.displot(train, x="target", bins=100, kde=True, height = 10)
```
